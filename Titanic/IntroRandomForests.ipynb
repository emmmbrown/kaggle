```python
# Model of Titanic survivors generated through Random Forests Machine Learning Algorithm
# Completed as an introductory exercise to Kaggle

import csv
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Load training data as a pandas dataframe
df = pd.read_csv('train.csv', header = 0)

df.head(n=10)
```

```python
# Check out which columns have data types that need to be omitted or changed to float
df.info()
```

```python
# We can omit the first column as the passenger ID is a random number and therefore would not
# add anything useful to our data analysis. So, we should convert 'Survived' and 'Pclass'
# into floats and save them in our new data frame, dfclean
dfclean = df[['Survived', 'Pclass']].astype(float)
dfclean.info()
```

```python
# We don't necessarily need the passenger names in our analysis, so they will be omitted from
# the clean data. Our next step is then to convert the 'Sex' variable to float 0s and 1s and
# add it to our clean data.
dfclean['Gender'] = df['Sex'].map({'female':0, 'male':1}).astype(float)
dfclean.info()
```

```python
dfclean.head(n=5)
```

```python
# Age is already a float variable so let's add it to the clean data.
dfclean['Age'] = df['Age'].astype(float)
dfclean.info()
```

```python
# Note that the number of entries is lower than the total 891 passengers we have data for.
# This is because the data is missing for the remaining passengers. To make up for this, let's
# make educated guesses as to their age based on the median from their gender and class. What we
# need to do is find the median ages and place them into a matrix.
median_ages = np.zeros((2, 3))
for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i, j] = dfclean[(dfclean['Gender'] == i) & \
                               (dfclean['Pclass'] == j+1)]['Age'].dropna().median()
median_ages
```

```python
# Next we copy the age column into a new column we will be working on. The loop will loop over the
# media ages matrix, and find all missing values with that class and gender, and replace the
# missing value with the median.
dfclean['AgeFill'] = dfclean['Age']
for i in range(0, 2):
    for j in range(0, 3):
        dfclean.loc[(dfclean.Age.isnull()) & (dfclean.Gender == i) & (dfclean.Pclass == j+1), \
               'AgeFill'] = median_ages[i, j]

# Let's look at what we'll be replacing some of the missing age values with
dfclean[ dfclean['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```

```python
# Now that we are done doing the work on this column, we can copy it to our original Age column
# and drop 'AgeFill'
dfclean['Age'] = dfclean['AgeFill']
dfclean = dfclean.drop('AgeFill', 1)
dfclean.info()
```

```python
# The variables SibSp and Parch are integers which need to be converted to floats as well:
dfclean['SibSp'] = df['SibSp'].astype(float)
dfclean['Parch'] = df['Parch'].astype(float)
dfclean.info()
```

```python
# Ticket number is also unnecessary data, so we'll skip that too. Next is fare, which is
# already a float value.
dfclean['Fare'] = df['Fare'].astype(float)
dfclean.info()
```

```python
# Now, there are a significant number of missing cabin values. Since it would require detailed
# information about the cabin layout of the ship to try to fill the missing values in, I will skip
# this column for now.
# Finally, we are at the 'Embarked' variable. Each letter represents one of three points of
# embarkation for each passenger.
# (C = Cherbourg; Q = Queenstown; S = Southampton) -> 1, 2, 3, NaN -> 0

dfclean['Embnumber'] = df['Embarked'].fillna(0)
dfclean['Embnumber'] = dfclean['Embnumber'].map({'C':1, 'Q':2, 'S':3, 0:0}).astype(float)

# Finally, we have our final, cleaned up data.
dfclean.info()
print dfclean.head(n=5)
```

```python
# To make it suitable for analysis, we need to save it as an array named train_data.
train_data = dfclean.values
train_data
```

```python
# Before we start to fit our model, we need to open the test data and perform the same
# operations on it that we did to our training data. Note the test data does not have the
# 'Survived' column because that is what we are trying to predict!
test_df = pd.read_csv('test.csv', header=0)
test_dfclean = test_df[['Pclass']].astype(float)
test_dfclean['Gender'] = test_df['Sex'].map({'female':0, 'male':1}).astype(float)
test_dfclean['Age'] = test_df['Age'].astype(float)
test_dfclean['AgeFill'] = test_dfclean['Age']
for i in range(0, 2):
    for j in range(0, 3):
        test_dfclean.loc[(test_dfclean.Age.isnull()) & (test_dfclean.Gender == i) & \
                         (test_dfclean.Pclass == j+1), 'AgeFill'] = median_ages[i, j]
test_dfclean['Age'] = test_dfclean['AgeFill']
test_dfclean = test_dfclean.drop('AgeFill', 1)
test_dfclean['SibSp'] = test_df['SibSp'].astype(float)
test_dfclean['Parch'] = test_df['Parch'].astype(float)
test_dfclean['Fare'] = test_df['Fare'].astype(float)
test_dfclean['Embnumber'] = test_df['Embarked'].fillna(0)
test_dfclean['Embnumber'] = test_dfclean['Embnumber'].map({'C':1, 'Q':2, 'S':3, 0:0}).astype(float)
test_dfclean.info()
print test_dfclean.head(n=5)
```

```python
# Note that there is a missing fare in the test data. We need to replace it with the median
# for that class.
if len(test_dfclean.Fare[ test_dfclean.Fare.isnull() ]) > 0:
    median_fare = np.zeros(3)
    for f in range(0,3):
        median_fare[f] = test_dfclean[ test_dfclean.Pclass == f+1 ]['Fare'].dropna().median()
    for f in range(0,3):
        test_dfclean.loc[ (test_dfclean.Fare.isnull()) & \
                         (test_dfclean.Pclass == f+1 ), 'Fare'] = median_fare[f]
test_dfclean.info()
```

```python
# Save the Passenger ID's for the data we will submit, and convert our final test data to
# a numpy array for the Random Forest!
ids = test_df['PassengerId'].values
test_data = test_dfclean.values

# Finally we can fit our data.
# Create the random forest object which will include all the parameters
# for the fit
forest = RandomForestClassifier(n_estimators = 100)

# Fit the training data to the Survived labels and create the decision trees
forest = forest.fit(train_data[0::,1::],train_data[0::,0])

# Take the same decision trees and run it on the test data
output = forest.predict(test_data).astype(int)

# Now let's open and save our prediction file so it's ready to submit to Kaggle!
predictions_file = open("myfirstforest.csv", "wb")
open_file_object = csv.writer(predictions_file)
open_file_object.writerow(["PassengerId","Survived"])
open_file_object.writerows(zip(ids, output))
predictions_file.close()
```
